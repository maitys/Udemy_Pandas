### 6. The `.agg()` Method

```python
fortune = pd.read_csv(filepath_or_buffer = 'fortune1000.csv', index_col="Rank")
fortune["Sector"] = fortune["Sector"].astype("category")
sectors = fortune.groupby(by="Sector")
fortune.head(n=3)
print("-" * 80)
fortune.info()
print("-" * 80)
print(type(fortune))
print(type(sectors))
```

```python
sectors.agg("mean").head(n=3)
```

```python
sectors.agg({"Revenue": "sum", 
             "Profits": "max", 
             "Employees": "mean"}).head(n=3)
```

```python
sectors.agg({"Revenue": "sum", 
             "Profits": ["max", "min"], 
             "Employees": "mean"}).head(n=3)
```

```python
aggregation = {"revenue_sum":pd.NamedAgg(column="Revenue", aggfunc="sum"),
               "profits_max":pd.NamedAgg(column="Profits", aggfunc="max"),
               "profits_min":pd.NamedAgg(column="Profits", aggfunc="min"),
               "employees_mean":pd.NamedAgg(column="Employees", aggfunc="mean")}
                                            
sectors.agg(**aggregation).head(n=3)
```

### 7. Iterating through Groups

```python
fortune = pd.read_csv(filepath_or_buffer = 'fortune1000.csv', index_col="Rank")
fortune["Sector"] = fortune["Sector"].astype("category")
sectors = fortune.groupby(by="Sector")
fortune.head(n=3)
print("-" * 80)
fortune.info()
print("-" * 80)
print(type(fortune))
print(type(sectors))
```

```python
sectors["Revenue"].max()
```

```python
df = pd.DataFrame(columns=fortune.columns)
df.head()
```

```python
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_company_in_group)
```

```python
df
```

```python
cities = fortune.groupby("Location")
```

```python
df = pd.DataFrame(columns=fortune.columns)
df.head()

for city, data in cities:
    highest_revenue_city_in_group = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_city_in_group)

df
```
