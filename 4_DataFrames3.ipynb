### 12. Filtering with the `where` Method

```python
bond = pd.read_csv(filepath_or_buffer = 'jamesbond.csv', 
                   index_col = "Film") # can use index_col to set a column as index
bond.sort_index(inplace = True)
bond.head(n=3)
```

```python
#Where returns all rows
filter1 = bond["Actor"] == "Sean Connery"
bond[filter1].head(n=3)
```

```python
#Where returns all rows but shows values for rows that True based on the condition
filter1 = bond["Actor"] == "Sean Connery"
bond.where(filter1).head(n=5)
```

```python
#Where returns all rows but shows values for rows that True based on the condition
bond.where(bond["Box Office"] > 500).head(n=3)
```

```python
filter1 = bond["Actor"] == "Sean Connery"
filter2 = bond["Box Office"] > 400
bond.where(filter1 & filter2).head(n=5)
```

### 13. The `.query()` Method

- Another way to filter dataframe -- Only when column names have no spaces in name
- More efficient way to filter. Not used often

```python
bond = pd.read_csv(filepath_or_buffer = 'jamesbond.csv', 
                   index_col = "Film") # can use index_col to set a column as index
bond.sort_index(inplace = True)
bond.head(n=3)
```

```python
#Another way to filter dataframe -- Only when column names have no spaces in name
bond.columns = ['Year', 'Actor', 'Director', 
                'Box_Office', 'Budget', 'Bond_Actor_Salary']
```

```python
bond.query("Actor == 'Sean Connery'").head(n=2)
print("-"*50)

bond.query("Director == 'Terence Young'").head(n=2)
print("-"*50)

bond.query("Actor != 'Roger Moore'").head(n=2)
print("-"*50)

bond.query("Actor == 'Sean Connery' and Box_Office > 400").head(n=2)
print("-"*50)

bond.query("Actor == 'Sean Connery' or Director == 'John Glen'").head(n=2)
print("-"*50)

bond.query("Actor in ['Timothy Dalton', 'George Lazenby']").head(n=2)
print("-"*50)

bond.query("Actor not in ['Sean Connery', 'Roger Moore']").head(n=2)
print("-"*50)
```

### 14. A Review of `.apply()` Method on Single Column

```python
bond = pd.read_csv(filepath_or_buffer = 'jamesbond.csv', 
                   index_col = "Film") # can use index_col to set a column as index
bond.sort_index(inplace = True)
bond.head(n=3)
```

```python
def convet_to_string_and_add_millions(number):
    return str(number) + ' Millions!'

bond["Box Office"]= bond["Box Office"].apply(convet_to_string_and_add_millions)
bond["Budget"]= bond["Budget"].apply(convet_to_string_and_add_millions)
bond["Bond Actor Salary"]= bond["Bond Actor Salary"].apply(convet_to_string_and_add_millions)
bond.head(n=3)
```

```python
bond = pd.read_csv(filepath_or_buffer = 'jamesbond.csv', 
                   index_col = "Film") # can use index_col to set a column as index
bond.sort_index(inplace = True)
bond.head(n=3)
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convet_to_string_and_add_millions)

bond.head(n=3)
```

### 15. The `.apply()` Method with rows

```python
bond = pd.read_csv(filepath_or_buffer = 'jamesbond.csv', 
                   index_col = "Film") # can use index_col to set a column as index
bond.sort_index(inplace = True)
bond.head(n=3)
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"

bond["status"] = bond.apply(good_movie, axis = 'columns') # 1 or 'columns': apply function to each row.
bond.head(n=3)
```
