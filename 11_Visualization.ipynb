### 4. Bar Graphs

```python
scor = data.DataReader(name="SCOR", data_source="yahoo", 
                     start=dt.date(2007,9,19), end=dt.datetime.now())
scor.head(n=3)
```

```python
def rank_performance(stock_price):
    if stock_price <= 20:
        return "Poor"
    elif stock_price > 20 and stock_price <= 30:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
scor["Close"].apply(rank_performance).value_counts()
```

```python
plt.style.use("ggplot")
holder = scor["Close"].apply(rank_performance).value_counts()
holder.plot(kind="bar", color="red", edgecolor="black");
```

```python
plt.style.use("ggplot")
holder = scor["Close"].apply(rank_performance).value_counts()
holder.plot(kind="barh", color="silver", edgecolor="black");
```

### 5. Pie Charts

```python
aapl = data.DataReader(name="AAPL", data_source="yahoo", 
                     start=dt.date(2009,9,19), end=dt.datetime.now())
aapl.head(n=3)
```

```python
aapl["Close"].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 108.57923588026925:
        return "Above Average"
    else:
        return "Below Average"
```

```python
plt.style.use("dark_background")
holder = aapl["Close"].apply(rank_performance).value_counts()
holder.plot(kind="pie", legend=True, figsize=(5,4));
```

### 6. Histograms

```python
goog = data.DataReader(name="GOOG", data_source="yahoo", 
                     start=dt.date(2004,8,19), end=dt.datetime.now())
goog.head(n=3)
```

```python
def custom_round(stock_price):
    return int(stock_price/100.0) * 100
```

```python
goog["High"].apply(custom_round).value_counts().sort_index()
```

```python
plt.style.use("ggplot")
holder = goog["High"].apply(custom_round).value_counts().sort_index()
holder.plot(kind="hist", bins= 15, edgecolor='black')
```
