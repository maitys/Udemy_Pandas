### 5. Outer Joins

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")

week1.head(n=2)
print("-" * 80)
week2.head(n=2)
print("-" * 80)
customers.head(n=2)
print("-" * 80)
foods.head(n=2)
```

```python
# Full outer join (A or B)
week1.merge(right=week2, how="outer", on="Customer ID", suffixes=["-Week1", "-Week2"])
```

```python
len(week1.merge(right=week2, how="outer", on="Customer ID", suffixes=["-Week1", "-Week2"]))
```

```python
# Full outer join (A or B)
week1.merge(right=week2, how="outer", on="Customer ID", suffixes=["-Week1", "-Week2"],
            indicator=True)
```

```python
# Full outer join (A or B)
merged_weeks = week1.merge(right=week2, how="outer", on="Customer ID", suffixes=["-Week1", "-Week2"],
                           indicator=True)
```

```python
merged_weeks["_merge"].value_counts()
```

```python
# A or B but not in A and B
filter1 = merged_weeks["_merge"].isin(["left_only", "right_only"])
merged_weeks[filter1]
```

```python
# A and B only (inner)
filter2 = merged_weeks["_merge"].isin(["both"])
merged_weeks[filter2]
```

### 8. Merging by Indexes with the `left_index` and `right_index` parameters

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")

week1.head(n=2)
print("-" * 80)
week2.head(n=2)
print("-" * 80)
customers.head(n=2)
print("-" * 80)
foods.head(n=2)
```

```python
customers = customers.set_index(keys="ID")
foods = foods.set_index(keys="Food ID")
customers.head(n=2)
print("-" * 80)
foods.head(n=2)
```

```python
# Join a column to an index. No duplicates
week1.merge(right=customers, how="left", left_on="Customer ID", right_index=True, sort=True)
sales = week1.merge(right=customers, how="left", left_on="Customer ID", right_index=True, sort=True)
```

```python
sales.merge(right=foods, how="left", left_on="Food ID", right_index=True, sort=True)
sales = sales.merge(right=foods, how="left", left_on="Food ID", right_index=True, sort=True)
```

```python
week1.merge(right=week2, how="left", left_index=True, right_index=True, sort=True, suffixes=["-Week1", "-Week2"])
```

### 9. The `join()` Method

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
satisfaction = pd.read_csv("Restaurant - Week 1 Satisfaction.csv")

week1.head(n=2)
print("-" * 80)
week2.head(n=2)
print("-" * 80)
customers.head(n=2)
print("-" * 80)
foods.head(n=2)
print("-" * 80)
satisfaction.head(n=2)
```

```python
week1.merge(right=satisfaction, how="left", left_index=True, right_index=True, sort=True)
```

```python
# used to concatenate vertically when dfs share same index
week1.join(other=satisfaction, how="left")
```
